```python
import numpy as np
import pandas as pd

# Set random seed for reproducibility
np.random.seed(42)

# 1. Original activations (10 units)
y = np.array([2, 3, 5, 4, 6, 1, 7, 3, 2, 5])

# 2. Dropout parameters
q = 0.8  # keep probability

# 3. Compute original variance across units
original_variance = np.var(y, ddof=0)
```

```python
num_samples = 10  # number of dropout samples
```

```python
# 4. Simulate inverted dropout
dropout_samples = []
for _ in range(num_samples):
    mask = np.random.binomial(1, q, size=y.shape)
    y_drop = (mask / q) * y
    dropout_samples.append(y_drop)
```

```python
dropout_samples
```

```python
# 5. Compute empirical variance per unit
empirical_variance = np.var(dropout_samples, axis=0, ddof=0)

# 6. Compute theoretical variance per unit
theoretical_variance = ((1 - q) / q) * y**2

# 7. Prepare a table
table = pd.DataFrame({
    "Unit": np.arange(1, len(y)+1),
    "Original Activation": y,
    "Empirical Variance": np.round(empirical_variance, 3),
    "Theoretical Variance": np.round(theoretical_variance, 3)
})

# 8. Print results
print(f"Original variance across units (before dropout): {original_variance:.3f}\n")
print(table)
```
